```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
train_review = pd.read_json('../data/restaurant/train_yelp_academic_dataset_review.json')
test_review = pd.read_json('../data/restaurant/test_yelp_academic_dataset_review.json')
user = pd.read_json('../data/restaurant/yelp_academic_dataset_user.json')
business = pd.read_json('../data/restaurant/yelp_academic_dataset_business.json')
```

```python
train_review.to_json('../data/restaurant/train_yelp_academic_dataset_review.json', orient='records', lines=True)
test_review.to_json('../data/restaurant/test_yelp_academic_dataset_review.json', orient='records', lines=True)
user.to_json('../data/restaurant/yelp_academic_dataset_user.json', orient='records', lines=True)
business.to_json('../data/restaurant/yelp_academic_dataset_business.json', orient='records', lines=True)
```

```python
def tokenize(row):
    words = str(row).split(',')
    words = [word.strip() for word in words]
    return words

user['friends'] = user['friends'].apply(tokenize)
```

```python
user = user.explode('friends')
```

```python
user = user[['user_id', 'friends']]
```

```python
user
```

```python
user = user[user['friends'] != 'None']
```

```python
user
```

```python
user_item = train_review[['business_id', 'user_id']]
```

```python
user_item = pd.merge(user_item, business, on='business_id', how='left')
```

```python
user_item = user_item[['business_id', 'user_id', 'attributes', 'category']]
```

```python
user = pd.merge(user_item, user, on='user_id', how='left', suffixes=('_left', ''))
```

```python
user = pd.merge(user, user_item, left_on='friends', right_on='user_id', suffixes=('_user', '_friend'))
```
